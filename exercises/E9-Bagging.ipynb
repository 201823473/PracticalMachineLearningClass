## Mashable news stories analysis

Predicting if a news story is going to be popular


```python
import pandas as pd

url = 'https://raw.githubusercontent.com/albahnsen/PracticalMachineLearningClass/master/datasets/mashable.csv'
df = pd.read_csv(url, index_col=0)
df.head()
```

```python
df.shape
```

```python
X = df.drop(['url', 'Popular'], axis=1)
y = df['Popular']
```

```python
y.mean()
```

```python
# train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
y_test.shape
```

# Exercise 9.1

Estimate a Decision Tree Classifier and a Logistic Regresion

Evaluate using the following metrics:
* Accuracy
* F1-Score

```python
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(random_state=1)
treeclf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
treeclf.score(X_test, y_test)
y_pred = treeclf.predict(X_test)
score=treeclf.score(X_test, y_test)
print('Score: \n',score)
accuracy = accuracy_score(y_test, y_pred, normalize=False)
print('Casos clasificados con precision: ' + str(accuracy))
```

```python
from sklearn.metrics import f1_score
y_pred = treeclf.predict(X_test) 
print('f1 score(default): \n',  f1_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
print('Score: \n',score)
y_pred = clf.predict(X_test) 
print('f1 score(default): \n',  f1_score(y_test, y_pred))
```

#### Mientras que para el árbol de decisión el F1 score es igual a 0.54 para la regresión logistica es 0.61, que interpretamos como el grado ponderado de clasificaciones correctas que hace respectivamente cada uno de los modelos, si apreciamos el score este no es ponderado y el resultado es el mismo para ambos modelos

# Exercise 9.5

Ensemble using weighted voting using the oob_error

Evaluate using the following metrics:
* Accuracy
* F1-Score

```python
models=[DecisionTreeClassifier(max_depth=None, random_state=123),DecisionTreeClassifier(max_depth=2, random_state=123),
        LogisticRegression(solver='liblinear',C=1e9, random_state=123),]
print(models)

y_pred_mean=[]
y_pred_df=[]
fs=[]
j=1
for model in list(models):
    clf = BaggingClassifier(base_estimator=model,n_estimators=100, bootstrap=True, random_state=42, n_jobs=-1, oob_score=True)
    clf.fit(X_train,y_train)
    errors=np.zeros(clf.n_estimators)
    y_pred_all_=np.zeros((X_test.shape[0],clf.n_estimators))
    for b in range(clf.n_estimators):
        oob_sample = ~clf.estimators_samples_[b]
        y_pred_=clf.estimators_[b].predict(pd.DataFrame(X_train).values[oob_sample])
        errors[b]=metrics.accuracy_score(y_pred_, y_train.values[oob_sample])
        y_pred_all_[:,b]=clf.estimators_[b].predict(X_test)
    alpha=(1-errors)/ (1-errors).sum()
    y_pred=(np.sum(y_pred_all_*alpha, axis=1)>=0.5).astype(np.int)
    y_pred_df.append([y_pred])
    y_pred_mean.append([b, y_pred.mean()])
    fs.append([j,round(f1_score(y_pred,y_test),3),round(clf.score(X_test,y_test),3)])
    j=j+1
y_pred_mean=pd.DataFrame(y_pred_mean)
y_pred_mean[1].tolist()
fs=pd.DataFrame(fs, columns=['prob','F1_score','tree_score'])
```

```python
print(fs)
```

```python
print(y_pred_df[1])
```

#### volviendo al criterio de 1 de un 50%, podemos encontrar que al incluir el concepto de OOB para mejorar las predcciones el F1 score crece en el caso del modelo de árboles de decisión a 64.5% y el de regresión logística a 65%
